```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
#Import the Advertising Data.
df = pd.read_csv("Advertising.csv", index_col = 0)
df.columns = ["TV","Radio","Newspaper","Sales"]
df.head()
```

```python
#Check for the null values in the column.
df.isna().sum()
```

```python
df.describe()
```

```python
df.info()
```

```python
#Check for the outliers in the dataset and univariate analysis of dataset.
for i in df.columns:
    plt.boxplot(df[i])
    plt.title("Plot of {}.".format(i))
    plt.show()
```

```python
#Bivariate Analysis of the dataset.
sns.pairplot(data = df, x_vars = ["TV","Radio","Newspaper"], y_vars = "Sales", height = 7, kind = 'reg')
```

```python
#Check for muticollinearity in the dataset.
sns.heatmap(data = df.corr(), annot = True, square = True)
```

```python
#Splitting of dependent and independent variable.
x = df.iloc[:,0:3]
y = df.iloc[:,-1]
```

```python
#Splitting the data into training and testing set.
x_train, x_test, y_train, y_test = train_test_split(x,y, random_state = 123)
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
#Fitting the linear model.
model = LinearRegression()
model.fit(x_train, y_train)
print(model.coef_)
print(model.intercept_)
```

```python
model.coef_
```

```python
#Predicting the test data.
y_pred = model.predict(x_test)
print(y_pred)
```

```python
#Check the model accuracy and fitting.
from sklearn.metrics import r2_score, mean_squared_error
rmse_train = np.sqrt(mean_squared_error(model.predict(x_train),y_train))
rmse_test = np.sqrt(mean_squared_error(y_pred,y_test))
print("RMSE of train data is {}.".format(rmse_train))
print("RMSE of test data is {}.".format(rmse_test))
print("The adjusted R square value of the training data is {}.".format(r2_score(model.predict(x_train),y_train)))
print("The adjusted R square value of the testing data is {}.".format(r2_score(y_pred,y_test)))
```
